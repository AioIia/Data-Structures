### Example 1: Time Complexity

What is the time complexity of access, search, insertion, deletion of array data structure? Explain why.


```python
import numpy as np

class MyArray:
    def __init__(self, input_list):
        self.array = np.array(input_list, dtype="i")

    def myInsert(self, x, i):
        new_array = np.empty(len(self.array) + 1, dtype="i")
        new_array[:len(self.array)] = self.array
        n = len(new_array)

        # Shift elements to the right
        for idx in range(n - 2, i - 1, -1):
            new_array[idx + 1] = new_array[idx]

        new_array[i] = x
        self.array = new_array  # Update self.array
        return self.array

    def myIndex(self, x):
        n = len(self.array)
        for idx in range(n):
            if self.array[idx] == x:
                return idx
        return -1  # If the key is not found

    def myDelete(self, i):
        n = len(self.array)

        # Shift elements to the left
        for idx in range(i + 1, n):
            self.array[idx - 1] = self.array[idx]

        self.array = self.array[:n - 1]  # Update self.array
        return self.array

    def myRemove(self, x):
        i = self.myIndex(x)
        if i == -1:
            print("Element not found.")
            return self.array

        print("Element found at position:", i)
        self.myDelete(i)
        return self.array
```

access: O(1), O(1) (direct access)
search: O(n) (last element), O(1) (first element)
insertion: O(n) (at the begining), Ω(1) (at the end)
deletion: O(n) (at the beginning), Ω(1) (at the end) 
removal: O(n), O(n) (you search then you remove)

### Example 4

Create a function that returns how many number of elements are odd and even elements in the array.


```python
def EvenOdd(arr):
    res = [0, 0]
    for el in arr:
        res[el % 2] += 1

    print(f"Even count: {res[0]}")
    print(f"Odd count: {res[1]}")

a = [1, 2, 2, 3, 4, 5, 5, 6, 7, 7]
EvenOdd(a)
```

Write a function `list_cumulated_sum(L, inf, sup)` that takes a non-empty list `L` and two integers inf and sup, where `inf < sup`. 

The function should return a new list containing the ith elements from `L` while the cumulative sum is ∈ `[inf, sup]`.

Example:

\>\>\> `list_cumulated_sum([3, 5, 4, 2, 1], 3, 10)`  
`[3, 5]`  
\>\>\> `list_cumulated_sum([1, 4, 2, 3], 15, 20)`  
`[]`

```python
def list_cumulated_sum(L, inf, sup):
    return []

print(list_cumulated_sum([3, 5, 4, 2, 1], 3, 10))
print(list_cumulated_sum([1, 4, 2, 3], 15, 20))
print(list_cumulated_sum([1, 2, 5, 4, 2, 1, 3, 5], 4, 20))
```

#### Solution

```python
def list_cumulated_sum(L, inf, sup):
    i, n = 0, len(L)
    sum, res = 0, []
    while i < n and sum + L[i] < inf:
        sum += L[i]
        i += 1
    while i < n and sum + L[i] < sup:
        sum += L[i]
        res.append(L[i])
        i += 1
    return res

print(list_cumulated_sum([3, 5, 4, 2, 1], 3, 10))
print(list_cumulated_sum([1, 4, 2, 3], 15, 20))
print(list_cumulated_sum([1, 2, 5, 4, 2, 1, 3, 5], 4, 20))
```

#### Visualization

```python
import matplotlib.pyplot as plt

def plot_cumulative_sum(L, inf, sup):
    cum_sum = 0
    cum_values = []
    colors = []
    stopped = False

    for val in L:
        if stopped:
            break

        if cum_sum + val > sup:
            cum_values.append(cum_sum + val)
            colors.append('#f44336')
            stopped = True
        else:
            cum_sum += val
            cum_values.append(cum_sum)
            if inf <= cum_sum <= sup:
                colors.append('#4CAF50')
            else:
                colors.append('#f44336')

    result = list_cumulated_sum(L, inf, sup)

    fig, ax = plt.subplots(figsize=(8, 5), dpi = 300)
    ax.bar(range(len(cum_values)), cum_values, color=colors)
    ax.axhline(inf, color='blue', linestyle='--', label=f'inf = {inf}')
    ax.axhline(sup, color='orange', linestyle='--', label=f'sup = {sup}')
    ax.set_xlabel('Index')
    ax.set_ylabel('Cumulative Sum')
    ax.set_title('Cumulative Sum Progression')
    ax.legend()

    plt.suptitle(f"Input: {L}", fontsize=12, y=1.02)
    plt.figtext(0.5, -0.1, f"Result: {result}", wrap=True, ha='center', fontsize=12)
    plt.tight_layout()
    plt.show()

plot_cumulative_sum([1, 2, 5, 4, 2, 1, 3, 5], 4, 20)
```
